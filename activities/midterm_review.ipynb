# I want to see everything that happened in 2010...

- I want a new dataset that only has rows for when `year == 2010`

```python
filtering_condition = df['year'] == 2010
df_2 = df.loc[filtering_condition, ['year', 'state', 'name', 'nonreligpct']]
```

```python
df_2
```

# We should find some statistics for our non-religious column in 2010 now
## We can find the mean and median using `.mean()` and `.median()`

```python
mean = df_2["nonreligpct"].mean()
median = df_2["nonreligpct"].median()
```

```python
print(f"The mean for Percentage of adherents to Non-Religious is: {mean*100}%.")
print(f"The median for Percentage of adherents to Non-Religious is: {median*100}%.")
```

# 3. Pre-processing

- We can remove duplicates
- We can remove null or empty values

```python
df_2.dropna(subset='nonreligpct', inplace=True)
```
